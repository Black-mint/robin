# Robin

```python
import os

# Edisonの代わりにローカル文献エージェントを使う
os.environ["ROBIN_LITERATURE_BACKEND"] = "local_agent"

# OpenAlexキーが不安定なら、まず匿名で試す
os.environ.pop("OPENALEX_API_KEY", None)

from robin.configuration import RobinConfiguration
from robin.assays import experimental_assay
from robin.candidates import therapeutic_candidates

# 情報工学分野の研究テーマとしてRobinを設定する
config = RobinConfiguration(
    application_domain="information_engineering",
    research_topic="CG crowds dataset for object detection",
    num_queries=5,
    num_assays=5,
    num_candidates=5,
)

print(config.application_domain)
print(config.disease_name)
print(os.getenv("ROBIN_LITERATURE_BACKEND"))
```

## Robin Set Up

```python
# ここでconfigを再作成すると、情報工学用の設定が上書きされるため無効化する
# config = RobinConfiguration(disease_name="dry age-related macular degeneration")

import logging

logger = logging.getLogger("robin")
logger.setLevel(logging.INFO)
```

# Initial therapeutics hypothesis generation

### Experimental assay generation

```python
candidate_generation_goal = await experimental_assay(configuration=config)
print(candidate_generation_goal)
```

### Therapeutic candidate generation

```python
await therapeutic_candidates(
    candidate_generation_goal=candidate_generation_goal,
    configuration=config,
)
```
